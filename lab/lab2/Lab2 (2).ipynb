```python
# 1.6 Make a subset that contains date, price, condition, lat, long and waterfront.
df_subset = df_correct[['date', 'price', 'condition', 'lat', 'long', 'waterfront']]
df_subset
```

### 2 Listing price

```python
# 2.1 Compute the average listing price
df_subset['price'].mean()
```

The average listing price is 540088.1418.

```python
# 2.2 Compute the average price in each of the “condition” category
df_subset.groupby('condition')['price'].mean()
```

This code returns the average price for each of the five "condition" categories when executed. The output indicates that the average price rises as the category of condition improves, from approximately 334,431(USD) for condition 1 to approximately 612,418(USD) for condition 5.This makes intuitive sense, as a home in better condition should command a higher price than one in worse condition.

```python
# 2.3 Print three most expensive listings
sorted_price = df_subset.sort_values('price', ascending=False)
sorted_price.head(3)
```

```python
# Print three most cheapest listings
sorted_price.tail(3)
```

```python
# 2.4 Create a single data frame that contains both 3 cheapest and 3 most expensive listings
sorted_price_df = pd.concat([sorted_price.head(3), sorted_price.tail(3)]).sort_values('price')
sorted_price_df
```

```python
# 2.5 Extract the most expensive listing for each “condition” category
cheapest_by_condition = df_subset.loc[df_subset.groupby('condition')['price'].idxmin()]
cheapest_by_condition
```

```python
# Extract the cheapest listing for each “condition” category 
expensive_by_condition = df_subset.loc[df_subset.groupby('condition')['price'].idxmax()]
expensive_by_condition
```

### 3 Proportions 

```python
# 3.1 Percentage of listings cheaper than $200,000
(df_subset['price'] < 200000).mean() * 100
```

There are 3.6367% of listing are cheaper than $200k.

```python
# Percentage of listings are over $1M
(df_subset['price'] > 1000000).mean() * 100
```

There are 6.7783% of listings are over $1M.

```python
# 3.2 Percentage of listings less than $1M and overlooking waterfront
((df_subset['price'] < 1000000) & (df_subset['waterfront'] == 1)).mean() * 100
```

The percentage of listings cost less than $1M and are overlooking waterfront is 0.268357%.

### 4 Plotting

```python
# 4.1 Create a scatter plot of longitude and latitude, colored by waterfront
fig, ax = plt.subplots(figsize=(10,8))
ax.scatter(df_subset.loc[df_subset['waterfront'] == 0, 'long'], df_subset.loc[df_subset['waterfront'] == 0, 'lat'], color='pink', alpha=0.5, label='Non-waterfront')
ax.scatter(df_subset.loc[df_subset['waterfront'] == 1, 'long'], df_subset.loc[df_subset['waterfront'] == 1, 'lat'], color='blue', alpha=0.5, label='Waterfront')
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
ax.set_title('longitude/latitude of the listings.')
ax.legend(loc='upper right')
plt.show()
```

#### 4.2
As seen on the map, it shows the shape of Seattle.The waterfront properties are near lakes and sea shore.This makes logic, as waterfront properties are typically more desirable and thus cost more. The non-waterfront properties are more equitably distributed across the region.
